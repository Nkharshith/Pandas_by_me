```python
#Indexing and sliceing
#Easy
#1.Display only the Name and Salary columns.
df[["Name","Salary"]]

#2.Display the employee present at index 5.
df.loc[5]

#3.Display the value present at row 3 and column Department.
df.loc[3,"Department"]

#4.Display the first 5 rows and first 3 columns.
df.iloc[0:5,0:3]

#Medium (3)

#5.Display rows with indexes [1, 4, 7] and only columns Name, Department, and Salary.
df.loc[[1,4,7],["Name","Department","Salary"]]

#6.Display the last 3 rows and only the Age and Salary columns.
df.loc[7:10,["Age","Salary"]]
df.iloc[-3:,[1,4]]

#7.Display all rows where only the first and last clumns are shown.
df.loc[0:9,["Name","Salary"]]
#hard
#8.Display rows 2–8 and columns from Age to Department using label-based selection.
df.loc[2:8,["Age","Gender","Department"]]

# #9.Return the employee name corresponding to the highest salary using indexing only.
df["Salary"].idxmax()
df.loc[8,"Name"]

df.loc[df["Salary"].idxmax(),"Name"]
# #10.Display a 4×3 subset starting from row index 3 and column index 1.
# df.iloc[3:7,1:4]
```

```python
#Filtering
#Easy
#1.Find all employees whose salary is greater than 60,000.
df.query("Salary>60000")

#2.Find all employees whose age is less than 25.
df[(df["Age"]<25)]
df.query("Age<25")
#3.Find all employees from the IT department.
df[(df["Department"] == "IT")]

#4 Find all female employees.
df[(df["Gender"] == "F")]

#Medium
#5.Find employees whose salary is between 50,000 and 65,000.
df[(df["Salary"] >= 50000) & (df["Salary"] <= 65000)]
df[df["Salary"].between(50000,60000)]

#6.Find employees whose age is greater than 23 and belong to the IT department.
df[(df["Age"]>23)&(df["Department"]=="IT")]

#7 Find employees whose department is either HR or Sales.
df[(df["Department"]=="HR")|(df["Department"]=="Sales")]
df.query("Department == 'HR' or Department == 'Sales'")
#Hard

#8.Find employees earning above the average salary.
avg_salary = df["Salary"].mean()
avg_salary

df[(df["Salary"]>avg_salary)]

#9 Find employees whose salary is greater than the average salary of their own department.
ad = df.groupby("Department")["Salary"].transform("mean")
df[(df["Salary"]>ad)]

#10.Find the top 3 highest-paid employees without using:
df.nlargest(3,"Salary")
```

```python
#Groupby
#Easy

#1.Find the average salary of each department.
df.groupby("Department")["Salary"].mean()

#2.Find the total salary paid by each department.
df.groupby("Department")["Salary"].sum()

#3.Find the number of employees in each department.
df.groupby("Department")["Name"].count()

#4.Find the maximum salary in each department.
df.groupby("Department")["Salary"].max()

#5.Find the minimum salary in each department.
df.groupby("Department")["Salary"].min()

#6 Find the average age in each department.
df.groupby("Department")["Age"].mean()

#7 Find the average salary for each gender.
df.groupby("Gender")["Salary"].mean()

#8 Find employee count by:
df.groupby(["Department","Gender"]).size()
df.groupby(["Department","Gender"])["Name"].count()
#9 For each department, calculate:
# '''Average Salary
# Maximum Salary
# Minimum Salary
# Employee Count
# in a single query'''

df.groupby("Department")["Salary"].agg(["mean","min","max","count"])

# #10.Find the department with the highest average salary.
dept_avg = df.groupby("Department")["Salary"].mean().idxmax()
dept_avg
# #11.Find which department contributes the highest percentage of total salary expense.
total_salary = df["Salary"].sum()
dept_salary = df.groupby("Department")["Salary"].sum()

percentage = (dept_salary/total_salary)*100
percentage
```

Characteristics of a Series
1. They are Homogeneous - store one type of data.
2. They support Vectorized operations.
3. They can handle missing values with NaN.
4. They have mutable values but immutable size.

```python
# Custom Indexing
s2 = pd.Series([23, 24, 25, 26], index = ["Adam", "Eve", "Charlie", "Bob"])
print(s2["Eve"]) # 24
print(s2["Bob"]) # 26
# Vectorized Operations
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
print(s1 + s2)
# Mutable Values but immutable size
s = pd.Series([1, 2, 3, 4, 5])
s[0] = 100
print(s)
changed_s = s.drop(1)
print(changed_s)
print(s)
```

Query String Rules:

1. We write the condition in a string.

2. We can use operators like and,or,not,==,!=, >, <, >=, <=, etc.
3. Use backticks for column names with spaces or symbols.
4. Use @ for Python variables.

Important - Rule of thumb:

• Direct filtering using df[...] - prefer using &/| for AND & OR.

• Inside query()

**Data Cleaning**

• Handle Missing Values ( NaN )

1.**df.isnull()** - Shows boolean DataFrame, True where NaN

2.**df.isnull().sum()** - Count of NaNs per column

3.**df.dropna()**-Drop rows with any NaN

4.**df.fillna(val)**-Fills NaN with a value
